```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_classes=4,n_features=6,n_informative=4,n_clusters_per_class=1,random_state=41)
```

```python
dfX=pd.DataFrame(X,columns=['car_1','car_2','car_3','car_4','car_5','car_6'])
```

```python
dfy=pd.DataFrame(y,columns=['label'])
```

```python
df=pd.concat([dfX,dfy],axis=1)
```

```python
# plots
plt.figure(figsize=(12,8))

plt.plot(X[y==3, 0], X[y==3, 1], "r^", label="3")
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="2")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="1")
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="0")
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df,hue='label',palette=sns.color_palette("hls", 4),kind='scatter')
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=1000,n_features=6,centers=4,random_state=41)
```

```python
dfX=pd.DataFrame(X,columns=['car_1','car_2','car_3','car_4','car_5','car_6'])
dfy=pd.DataFrame(y,columns=['label'])
df=pd.concat([dfX,dfy],axis=1)
```

```python
# plots
plt.figure(figsize=(12,8))
plt.plot(X[y==3, 0], X[y==3, 1], "r^", label="3")
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="2")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="1")
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="0")
plt.legend()
plt.show()
```

```python
sns.pairplot(df,hue='label',palette=sns.color_palette("hls", 4),kind='scatter')
```
