```python
def categories_ages(age):
    if age <18:
        return 'mineur'
    elif (age >=18) & (age<=56):
        return 'adulte'
    elif (age >56) & (age <65):
        return 'senior'
    else:
        return ('vieux')
        
```

```python
data['age'].map(categories_ages)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y,label='quadratique')
plt.plot(x,x**3,label='cubique')
plt.title('figure 1')
plt.xlabel('Ox')
plt.ylabel('Oy')
plt.legend()
plt.show()

plt.figure()
plt.plot(x,x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,c='red')
plt.title('figure 1')
plt.subplot(2,1,2)
plt.plot(x,np.sin(x),label='sinus')
plt.plot(x,np.cos(x),label='cosinus')
plt.legend()
```

```python
#la méthode orientée objet existe et est plus compliquée !
fig,ax=plt.subplots(2,1,sharex=True) #on crée deux objets, la figure et les axes surl lesquels on veut travailler
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
noms=list(iris.target_names)
```

```python
x
```

```python
y
```

```python
noms
```

## SCATTER

```python
plt.scatter(x[:,2],x[:,3],c=y,alpha=0.5,s=100*x[:,2]) #alpha = transparence, s=taille des points
plt.xlabel('longueur pétale')
plt.ylabel('largeur pétale')
plt.legend(noms)
```

## Graphes 3D

```python
#les graphes 3D
#%matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax=plt.axes(projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],c=y)
```

```python
f=lambda x,y:np.sin(x)+np.cos(x+y)
```

```python
X=np.linspace(0,5,100)
Y=np.linspace(0,5,100)
X,Y=np.meshgrid(X,Y)

Z=f(X,Y)
```

```python
Z.shape
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma') #https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
```
